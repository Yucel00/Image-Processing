```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
my_image=cv2.imread("sara.jpg")
my_image_gray=cv2.cvtColor(my_image,cv2.COLOR_BGR2GRAY)
my_image_gray
```

### Adaptif esitleme islemleri

```python
"""
adaptif esitleme yontemi turu
adptif esikleme isleminde kullanilacak(baz alinacak) normal esikleme turu
255 en buyuk piksel degeri esiklemeden sonra verilecek  en buyuk pikseli degeri
11=Blok buyuklgu(11 satir 11 stunluk ) bir matris olan goruntu parcalara ayrilacak ve her parca icersinde lokal esikleme yapacak
"""
my_adaptive_image=cv2.adaptiveThreshold(my_image_gray,255,
                                       cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8,
                                       ) #adaptive goruntuyu pixelle islemek yerine bloklara ayirir 
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image_gray,cmap="gray")
plt.title("Gray")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(my_adaptive_image,cmap="gray")
plt.title("Adaptive")
plt.axis("off")
```

```python
my_adaptive_binary_inv_image=cv2.adaptiveThreshold(my_image_gray,255,
                                       cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,11,8,
                                       )
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.imshow(my_image_gray,cmap="gray")
plt.title("Gray")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(my_adaptive_image,cmap="gray")
plt.title("Adaptive")
plt.axis("off")
fig.add_subplot(rows,columns,3)
plt.imshow(my_adaptive_binary_inv_image,cmap="gray")
plt.title("Inv")
plt.axis("off")
```

```python
my_adaptive_binary_gaus_image=cv2.adaptiveThreshold(my_image_gray,255,
                                       cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8,
                                       )
```

```python
fig=plt.figure(figsize=(10,7))
rows=2
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image_gray,cmap="gray")
plt.title("Gray")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(my_adaptive_image,cmap="gray")
plt.title("Adaptive")
plt.axis("off")
fig.add_subplot(rows,columns,3)
plt.imshow(my_adaptive_binary_inv_image,cmap="gray")
plt.title("Inv")
plt.axis("off")
fig.add_subplot(rows,columns,4)
plt.imshow(my_adaptive_binary_gaus_image,cmap="gray")
plt.title("Gaus")
plt.axis("off")
```
