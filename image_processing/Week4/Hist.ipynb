```python
import numpy
import cv2
import matplotlib.pyplot as plt
```

```python
my_image_gray=cv2.imread("./cameraman.png",0)
img_eq_hist=cv2.equalizeHist(my_image_gray)
histr=cv2.calcHist([my_image_gray],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
my_image_gray.shape
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image_gray,cmap="gray")
plt.title("Gray")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(img_eq_hist,cmap="gray")
plt.title("Adaptive")
plt.axis("off")
```

```python
hist_org=cv2.calcHist([my_image_gray],[0],None,[256],[0,256])
hist_eq=cv2.calcHist([img_eq_hist],[0],None,[256],[0,256])

plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.plot(hist_org,color="b")
plt.title("Orjinal Goruntu Histogrami")
```
