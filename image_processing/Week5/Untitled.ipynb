```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
my_image=cv2.imread("./binary.png",0)
my_kernel=np.ones((5,5),dtype=np.uint8)
my_kernel
```

```python
my_erosion_img=cv2.erode(my_image,my_kernel,iterations=2) #erosion and kullanarak beyazlari kucultur ercode fonk kulanir
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.title("Erosion")
plt.axis("off")
```

```python
my_dilation_img=cv2.dilate(my_image,my_kernel,iterations=2) #dilation xor yaparak beyzlari genisletir dilate fonk kullandik
fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.imshow(my_image,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.title("Erosion")
plt.axis("off")
fig.add_subplot(rows,columns,3)
plt.imshow(my_dilation_img,cmap="gray")
plt.title("Dilation")
plt.axis("off")
```

```python
my_image2=cv2.imread("./binary_with_text.png",0)
white_noise=np.random.randint(0,2,size=my_image2.shape[:2])
white_noise*=255
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image2,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(white_noise,cmap="gray")
plt.title("White Noise")
plt.axis("off")
```

```python
my_noised_image=my_image2+white_noise
plt.imshow(my_noised_image,cmap="gray")
```

```python
my_opening_img=cv2.morphologyEx(my_noised_image.astype(np.float32),

                                cv2.MORPH_OPEN,my_kernel)
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=4
fig.add_subplot(rows,columns,1)
plt.imshow(my_image2,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(white_noise,cmap="gray")
plt.title("White Noise")
plt.axis("off")
fig.add_subplot(rows,columns,3)
plt.imshow(my_noised_image,cmap="gray")
plt.title("Noised")
plt.axis("off")
fig.add_subplot(rows,columns,4)
plt.imshow(my_opening_img,cmap="gray")
plt.title("Opening")
plt.axis("off")
```

```python
black_noise=np.random.randint(0,2,size=my_image2.shape[:2])
black_noise*=-255
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=5
fig.add_subplot(rows,columns,1)
plt.imshow(my_image2,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(white_noise,cmap="gray")
plt.title("White Noise")
plt.axis("off")
fig.add_subplot(rows,columns,3)
plt.imshow(my_noised_image,cmap="gray")
plt.title("Noised")
plt.axis("off")
fig.add_subplot(rows,columns,4)
plt.imshow(my_opening_img,cmap="gray")
plt.title("Opening")
plt.axis("off")
fig.add_subplot(rows,columns,5)
plt.imshow(black_noise,cmap="gray")
plt.title("Blck Noise")
plt.axis("off")
```

```python
my_black_noise_image=my_image2+black_noise
plt.imshow(my_black_noise_image,cmap="gray")
```

```python
 my_black_noise_image[my_black_noise_image<= -245]=0
```

```python
my_closing_img=cv2.morphologyEx(my_black_noise_image.astype(np.float32),

                                cv2.MORPH_CLOSE,my_kernel)
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=4
fig.add_subplot(rows,columns,1)
plt.imshow(my_image2,cmap="gray")
plt.title("Orginal Image")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(black_noise,cmap="gray")
plt.title("Black Noise")
plt.axis("off")
fig.add_subplot(rows,columns,3)
plt.imshow(my_black_noise_image,cmap="gray")
plt.title("My Black Noise")
plt.axis("off")
fig.add_subplot(rows,columns,4)
plt.imshow(my_closing_img,cmap="gray")
plt.title("Closing")
plt.axis("off")
```

```python
my_gradient_img_1=cv2.morphologyEx(my_image,cv2.MORPH_GRADIENT,my_kernel)#fonksiyon gicicek resmin renk degistirdigi yerleri cizicek geri kalai yok sayicak
my_gradient_img_2=cv2.morphologyEx(my_image2,cv2.MORPH_GRADIENT,my_kernel)#fonksiyon gelicek resmin bi anda degistigi yeri bulup cizicek
```
