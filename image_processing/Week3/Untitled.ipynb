```python
import cv2
import matplotlib.pyplot as plt
```

```python
my_image=cv2.imread("nebula.jpg")
my_img2=cv2.cvtColor(my_image,cv2.COLOR_BGR2GRAY)
my_image
```

```python
cv2.imshow("my_img2",my_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(my_image.shape)
print(my_img2.shape)
```

```python
__, my_binary_ug_thresh_img=cv2.threshold(my_img2,thresh=125,maxval=255 ,type=cv2.THRESH_BINARY)
```

```python
print(__)
```

```python
plt.imshow(my_binary_ug_thresh_img,cmap="gray")
plt.axis("off")
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_img2,cmap="gray")
plt.axis("off")
fig.add_subplot(rows,columns,2)
plt.imshow(my_binary_ug_thresh_img,cmap="gray")
plt.axis("off")
```

```python
__, my_binary_inv_thresh_img=cv2.threshold(my_img2,thresh=125,maxval=255 ,type=cv2.THRESH_BINARY_INV)

fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.imshow(my_img2)
plt.axis("off")
plt.title("Orginal")
fig.add_subplot(rows,columns,2)
plt.imshow(my_binary_inv_thresh_img,cmap="gray")
plt.axis("off")
plt.title("2.Resim")
fig.add_subplot(rows,columns,3)
plt.imshow(my_binary_inv_thresh_img,cmap="gray")
plt.axis("off")
plt.title("3.Resim")
```

```python
__, my_binary_trunc_thresh_img=cv2.threshold(my_img2,thresh=125,maxval=255 ,type=cv2.THRESH_TRUNC)
```

```python
fig=plt.figure(figsize=(10,7))
rows=2
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image)
plt.axis("off")
plt.title("Orginal")
fig.add_subplot(rows,columns,2)
plt.imshow(my_img2,cmap="gray")
plt.axis("off")
plt.title("Thresh")
fig.add_subplot(rows,columns,3)
plt.imshow(my_binary_inv_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Inv")

fig.add_subplot(rows,columns,4)
plt.imshow(my_binary_trunc_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Trunc")
```

```python
__, my_binary_tozero_thresh_img=cv2.threshold(my_img2,thresh=125,maxval=255 ,type=cv2.THRESH_TOZERO)
__, my_binary_tozeroinv_thresh_img=cv2.threshold(my_img2,thresh=125,maxval=255 ,type=cv2.THRESH_TOZERO_INV)
```

```python
fig=plt.figure(figsize=(10,7))
rows=3
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_image)
plt.axis("off")
plt.title("Orginal")
fig.add_subplot(rows,columns,2)
plt.imshow(my_img2,cmap="gray")
plt.axis("off")
plt.title("Thresh")
fig.add_subplot(rows,columns,3)
plt.imshow(my_binary_inv_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Inv")

fig.add_subplot(rows,columns,4)
plt.imshow(my_binary_trunc_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Trunc")

fig.add_subplot(rows,columns,5)
plt.imshow(my_binary_tozero_thresh_img,cmap="gray")
plt.axis("off")
plt.title("Tozero")
fig.add_subplot(rows,columns,6)
plt.imshow(my_binary_tozeroinv_thresh_img	,cmap="gray")
plt.axis("off")
plt.title("INV")
```
